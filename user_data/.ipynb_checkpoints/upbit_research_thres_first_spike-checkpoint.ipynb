```python
# classify symnbols
krw_symbols = [s for s in market_symbols if 'KRW' in s]
btc_symbols = [s for s in market_symbols if '/BTC' in s]
usdt_symbols = [s for s in market_symbols if '/USDT' in s] # not even usdt/krw pair

coin_pair_dict = {} # coin : available pairs

for k in krw_symbols:
    coin = k[:k.find('/')]
    if coin not in coin_pair_dict.keys():
        coin_pair_dict[coin] = [k]
    else:
        coin_pair_dict[coin].append(k)

for k in btc_symbols:
    coin = k[:k.find('/')]
    if coin not in coin_pair_dict.keys():
        coin_pair_dict[coin] = [k]
    else:
        coin_pair_dict[coin].append(k)

for k in usdt_symbols:
    coin = k[:k.find('/')]
    if coin not in coin_pair_dict.keys():
        coin_pair_dict[coin] = [k]
    else:
        coin_pair_dict[coin].append(k)

coin_single_pair = [] # coin: preferred pair

for k in coin_pair_dict.keys():
    coin_single_pair.append(coin_pair_dict[k][0])
    
    
coin_single_pair_dict = {}

for k in coin_pair_dict.keys():
    coin_single_pair_dict[k]=(coin_pair_dict[k][0])

upbit_tradable_coin = pd.read_csv('upbit_global.csv')['Unnamed: 1'].values.tolist() # krw

upbit_tradable_pair = []

for k in upbit_tradable_coin:
    upbit_tradable_pair.append(coin_single_pair_dict[k])  # krw pair`
#import json
#print(json.dumps(coin_single_pair))

krw_tradable_pair = [p for p in upbit_tradable_pair if 'KRW' in p]
btc_tradable_pair = [p for p in upbit_tradable_pair if 'BTC' in p]

#len(upbit_tradable_pair) # 128 pairs that are gloablly tradable 
```

```python
# get all data of these pairs
def file_name(pair, freq):
    name = f"{pair.replace('/', '_')}-{freq}.json"
    return name

def convert_unix_to_datetime(x):
    return datetime.utcfromtimestamp(int(x/1000))

def read_ochlv_json(pair, freq):
    raw_data_df = pd.read_json(f"upbit/{file_name(pair, freq)}")
    header = ['unix_time', 'open', 'high', 'low', 'close', 'volume']
    raw_data_df.columns = header
    raw_data_df['date'] = raw_data_df['unix_time'].apply(lambda x: convert_unix_to_datetime(x))
    #raw_data_df['pair'] = pair
    #raw_data_df = raw_data_df.set_index(['pair','time'])
    return raw_data_df
```
