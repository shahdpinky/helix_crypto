```python
from data_provider.glassnode_enums import *
from enums import Asset
from datetime import date, datetime
import pandas as pd
import numpy as np
```

```python
from data_provider.glassnode import GlassNodeKey, GlassNodeProvider
```

```python
price_key = GlassNodeKey(MarketGN.price, Asset.BTC, frequency='24h', currency='USD')
gn_pvd = GlassNodeProvider()
```

```python
start_date = date(2015, 1, 1)
end_date = date.today()
bitcoin_price_df = gn_pvd.get(price_key, start_date, end_date)
```

```python
stock_to_flow_deflection_key = GlassNodeKey(IndicatorsGN.stock_to_flow_deflection, Asset.BTC, frequency='24h') # ask why 1h can't
```

```python
stock_to_flow_deflection_df = gn_pvd.get(stock_to_flow_deflection_key, start_date, end_date)
```

```python
stock_to_flow_deflection_df
```

```python
combined_df = bitcoin_price_df.merge(stock_to_flow_deflection_df, left_index=True,right_index=True)
```

```python
combined_df['stock_to_flow_price'] = combined_df['market_price_usd_close']/combined_df['indicators_stock_to_flow_deflection']
```

```python
combined_df['log_price'] = np.log10(combined_df['market_price_usd_close'])
combined_df['log_stock_to_flow_price'] = np.log10(combined_df['stock_to_flow_price'])
```

```python
combined_df['log_stock_to_flow_delect'] = combined_df['log_price'] - combined_df['log_stock_to_flow_price'] 
```

```python
import plotly.express as px
fig = px.histogram(combined_df, x=['indicators_stock_to_flow_deflection', 'log_stock_to_flow_delect'])
fig.show()
```

```python
# my indicator is more normal
```

```python
fig = px.histogram(combined_df, x=['log_stock_to_flow_delect'])
fig.show()
```

```python
# caucious if model or factor is not smooth
px.line(combined_df, x=combined_df.index, y=['log_stock_to_flow_delect', 'log_stock_to_flow_price', 'log_price'])
```

```python
combined_df['price_return'] = combined_df['market_price_usd_close'].pct_change()
```

```python
fig = px.histogram(combined_df, x=['log_stock_to_flow_delect', 'price_return'])
fig.show()
```

```python
combined_df['log_stock_to_flow_delect_ema'] = combined_df['log_stock_to_flow_delect'].ewm(span=28).mean()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_regressor = LinearRegression()  # create object for the class
```

```python
combined_df['log_stock_to_flow_delect_ema'].values[:-1]
```

```python
combined_df['price_return'].values[1:]
```

```python
X = combined_df['log_stock_to_flow_delect_ema'].values[:-1].reshape(-1, 1)
Y = combined_df['price_return'].values[1:].reshape(-1, 1)
```

```python
linear_regressor.fit(X, Y)  # perform linear regression
Y_pred = linear_regressor.predict(X)  # make predictions
```

```python
import matplotlib.pyplot as plt  # To visuali
```

```python
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.show()
```

```python
linear_regressor.score(X, Y) # nothing to do with daily return 
```
